```python
import panda as pd
from bs4 import BeautifulSoup
import requests
import re

r  = requests.get("http://www.ohchr.org/EN/HRBodies/UPR/Pages/Documentation.aspx")

data = r.text

soup = BeautifulSoup(data, "lxml")

links = []

for link in soup.findAll('a', attrs={'href': re.compile("^http://")}):
    links.append(link.get('href'))

J = len("http://www.ohchr.org/EN/HRBodies/UPR/Pages/")
K = len("http://www.ohchr.org/EN/HRBodies/UPR/Pages/AFIndex.aspx")

link2 = [s for s in links if len(s) == K]

country_id = [x[J:J+2] for x in link2]

country_id
```
