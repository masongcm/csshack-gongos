## UK Data

```python
uk_orgs = ohchr_scrape(countries['UK'])
uk_orgs = get_pdfs(uk_orgs, folderpath = 'pdf/uk/')
uk_orgs_text = get_text(uk_orgs)
```

```python
text_p = []
for index, row in uk_orgs_text.iterrows():
    text_p.append(preprocess(row["text"], language=row["language"])[0])
uk_orgs_text['text_p'] = text_p
```

```python
uk_orgs_text.to_csv('../data/uk/uk_ngos_stems.csv')
```

## Cuba Data

```python
cu_orgs = ohchr_scrape(countries['Cuba'])
cu_orgs = get_pdfs(cu_orgs, folderpath = 'pdf/cu/')
cu_orgs_text = get_text(cu_orgs)
```

```python
# remove other languages
cu_orgs_text = cu_orgs_text.drop(cu_orgs_text[cu_orgs_text['language'] == 'русский'].index)
cu_orgs_text = cu_orgs_text.drop(cu_orgs_text[cu_orgs_text['language'] == '\n'].index)
cu_orgs_text = cu_orgs_text.drop(cu_orgs_text[cu_orgs_text['language'] == 'العربية'].index)

cu_orgs_text['language'].unique()

text_p = []
for index, row in cu_orgs_text.iterrows():
    text_p.append(preprocess(row["text"], language=row["language"])[0])
cu_orgs_text['text_p'] = text_p
```

```python
cu_orgs_text.to_csv('../data/uk/uk_ngos_stems.csv')
```
