```python
def extract_to_dict(country):
    text_dict = {}
    for index, row in country.iterrows():
        try: 
            text_dict[index] = [row['name'],[extract_text(row['file'])], row['lang']]
        except:
            print('Cannot extract text for ' + str(row['name']))
    return text_dict


data_dict = extract_to_dict(vz_csv)
```
